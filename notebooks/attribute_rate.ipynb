```python
import wikipedia
import yake
kw_extractor = yake.KeywordExtractor(lan="en", n=1, dedupLim=0.9, windowsSize=1, top=50)
```

```python
page = wikipedia.page("Sourav Ganguly")

print(page.content)
```

```python
keywords = kw_extractor.extract_keywords(page.content)
keywords
```
